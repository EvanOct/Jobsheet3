```python
class Student:
    def __init__(self, name, score):
        # Inisialisasi atribut privat
        self.__name = name
        self.__score = None
        self.__grade = None
        
        # Menggunakan setter agar validasi score dan penentuan grade berjalan
        self.score = score

    # Getter dan Setter untuk name
    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, value):
        self.__name = value

    # Getter dan Setter untuk score
    @property
    def score(self):
        return self.__score

    @score.setter
    def score(self, value):
        if 0 <= value <= 100:
            self.__score = value
            self.__update_grade()  # Update grade otomatis saat score berubah
        else:
            print("Kesalahan: Nilai harus berada di antara 0 hingga 100.")

    # Method internal untuk update grade secara otomatis
    def __update_grade(self):
        if self.__score >= 90:
            self.__grade = 'A'
        elif self.__score >= 80:
            self.__grade = 'B'
        elif self.__score >= 70:
            self.__grade = 'C'
        elif self.__score >= 60:
            self.__grade = 'D'
        else:
            self.__grade = 'E'

    # Method untuk menampilkan informasi lengkap
    def show_info(self):
        print(f"Nama Mahasiswa: {self.__name}")
        print(f"Nilai: {self.__score}")
        print(f"Grade: {self.__grade}")

    # Destruktor
    def __del__(self):
        print(f"Data mahasiswa {self.__name} telah dihapus dari sistem.")

# --- Pengujian Program (Sesuai Contoh Output) ---
if __name__ == "__main__":
    # Inisialisasi objek pertama
    mhs = Student("Siti", 87)
    mhs.show_info()

    print("\nNilai diubah...")
    mhs.score = 93 # Mengubah nilai (memicu update grade)
    mhs.show_info()
    print()

    # Menghapus objek untuk memicu destruktor
    del mhs
```
